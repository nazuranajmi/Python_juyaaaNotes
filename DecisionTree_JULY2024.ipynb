```python
import pandas as pd
import numpy as np

df = pd.read_csv("drug200.csv")

df.head()
```

```python
df.info()
```

```python
df['Drug'].value_counts()
```

```python
X = df.iloc[:, :-1].values
X
```

```python
y = df['Drug']
y
```

```python
X.head()
```

```python
#convert categorical to numeric

from sklearn.preprocessing import LabelEncoder

label_sex = LabelEncoder()

X[:,1] = label_sex.fit_transform(X[:, 1])

label_BP = LabelEncoder()

X[:,2] = label_BP.fit_transform(X[:, 2])

label_chole = LabelEncoder()

X[:,3] = label_chole.fit_transform(X[:, 3])

```

```python
X[:5]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)

X_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import jaccard_score

DS = DecisionTreeClassifier()

#training
DS.fit(X_train, y_train)

#prediction
y_hat =DS.predict(X_test)

#evaluation
print('The accuracy score is: ', accuracy_score(y_test, y_hat))
print('')
print('The confusion matrix is: ', confusion_matrix(y_test, y_hat))
print('')
print('The Jaccard Score is: ', jaccard_score(y_test, y_hat, average='micro'))

```
