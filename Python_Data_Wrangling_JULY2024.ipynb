```python
#min - max

mydata['length'] = (mydata['length'] - mydata['length'].min())/(mydata['length'].max() - mydata['length'].min())

mydata[['length','width','height']].head()
```

```python
#z-score

mydata['height'] = (mydata['height'] - mydata['height'].mean())/mydata['height'].std()

mydata[['length','width','height']].head()
```

```python
import numpy as np

msk = np.random.rand(len(mydata)) < 0.8

train = mydata[msk]
test = mydata[~msk]
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()

train_x = np.asanyarray(train[['highway_L/100KM']])
train_y = np.asanyarray(train[['price']])

#training/learning

regr.fit(train_x, train_y)
```

```python
print ('The value of the slop is: ', regr.coef_)
print ('The value of the intercept is: ', regr.intercept_)
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['highway_L/100KM']])
test_y = np.asanyarray(test[['price']])

test_y_ = regr.predict(test_x)

print('The R2 is: ', r2_score(test_y, test_y_))
```

```python
#highway_L/100KM vs price

regr1 = linear_model.LinearRegression()

train_x = np.asanyarray(train[['highway_L/100KM']])
train_y = np.asanyarray(train[['price']])

#training/learning

regr1.fit(train_x, train_y)

#prediction

test_x = np.asanyarray(test[['highway_L/100KM']])
test_y = np.asanyarray(test[['price']])

test_y_ = regr1.predict(test_x)

#evaluation
print ('The R2 is: ', round(((r2_score(test_y, test_y_))*100), 2))
print ('The MSE is: ', round(((np.mean((test_y_ - test_y)**2))*100), 2))
print ('The MAE is: ', round(((np.mean(np.absolute(test_y_ - test_y)))*100), 2))
```
